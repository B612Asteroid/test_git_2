```python
import csv
######
#. 파일을 읽습니다.
file = open("D:\data\경상남도 통영시_꿈이랑도서관보유도서목록현황_20230206.csv", 
            'r',
           encoding = 'UTF-8')

#. csv 라이브러리가  csv 파일을 읽을 수 있도록 변환합니다.
reader = csv.reader(file)

#  csv 파일의 한 라인(줄)씩 읽어 들이는 과정입니다.
cnt = 1
for line in reader :
    print(line)
    if cnt > 30 :
        break
    cnt = cnt + 1
```
