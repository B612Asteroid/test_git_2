```python
import csv
from datetime import datetime

#. 파일을 읽습니다.
file = open("D:\data\경상남도_통영시_통영시립욕지도서관보유도서목록현황_20230208.csv", 
            'r',
           encoding = 'UTF-8')

#. csv 라이브러리가  csv 파일을 읽을 수 있도록 변환합니다.
reader = csv.reader(file)

# 도서관명, 등록번호, 자료명, 저자명, 발행처,발행연도,관리기관명, 데이터기준일자

is_header = True
books = []
datetime_format = "%Y-%m-%d"

for line in reader:
    if is_header == True:
        is_header = False
        continue
    
    if len(line) < 8 :
        continue
        
    book = {
        "library_name" : line[0], 
        "book_number": int(line[1]), 
        "book_name": line[2],
        "author" : line[3],
        "publisher": line[4],
        "publish_year": line[5],
        "sido" : line[6],
        "create_date": datetime.strptime(line[7], datetime_format)
    }
    books.append(book)
    
len(books)
```
